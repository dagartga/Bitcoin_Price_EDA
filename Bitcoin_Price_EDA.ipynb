```python
# create a separate dataframe for just the price
price_df = btc_df[['priceUSD']]
price_df.head()
```

```python
import re

# create a regex for finding all features with price in the name
regex = re.compile(r'price\w+')
# create a list to append them to
price_list = []

for feat_name in btc_df.columns:
    match = re.findall(regex, feat_name)
    if len(match) > 0:
        price_list.append(match[0])

price_list
```

```python
# remove those columns from the dataframe
feat_df = btc_df.drop(price_list, axis=1)

# check that they are dropped
'price90rocUSD' in feat_df.columns, 'price14momUSD' in feat_df.columns
```

## Create many different future price values

Create a dataframe for the 1-day, 2-day, 3-day, 7-day, 14-day, and 30-day price

This will be something to search for correlation against for any features that might give insight to future price movement

```python
# create a function to return a series of future prices
def future_price(df, interval):
    """
        Takes in a df of only the price and the future price interval
        Returns a dataframe with the values for the prices from that interval
        Example: future_price(price_df, 1) returns the prices for one-day from actual date 
    """
    col_name = f'price_{interval}day'
    
    new_df = df[interval:]
    # change the column name
    new_df.columns = [col_name]
    # change the index to the original index of dates minus the last few rows
    new_df.index = df.index[:-interval]
    
    return new_df
    
```

```python
# create the different dataframes
price_1day = future_price(scaled_price_df, 1)
price_2day = future_price(scaled_price_df, 2)
price_3day = future_price(scaled_price_df, 3)
price_7day = future_price(scaled_price_df, 7)
price_14day = future_price(scaled_price_df, 14)
price_30day = future_price(scaled_price_df, 30)
```

### Create functions simplify the process

```python
# create a function to return sorted correlation with the various intervals
def price_corr(feat_df, price_df, interval):
    
    # join the dataframes
    df_joined = price_df.join(feat_df)
    
    # get correlations
    corr_matrix = df_joined.corr().round(2)
    
    # sort the correlation values
    sorted_corr = corr_matrix[f'price_{interval}day'].sort_values(ascending=False)
    
    return sorted_corr
```

```python
# create a function to remove the similar features
def simplified_corr(sorted_corr):
    
    # create a regex to only capture the beginning of the feature
    regex = re.compile(r'top100|[a-z_]+')

    # create a dictionary for correlations
    feat_corr_dic = {}
    # create a list for the full feature name
    full_feat_name = []

    # iterate through the correlation list and store only the highest correlation for each
    for feat, val in sorted_corr.items():
        match = re.search(regex, feat)
        name = match.group(0)
        if name not in feat_corr_dic:
            feat_corr_dic[name] = val
            full_feat_name.append(feat)
    
    return feat_corr_dic, full_feat_name
```

## Find Correlations for 2, 3, 7, 14, and 30-day Prices vs Features

### 2-Day Price Correlations

```python
corr_2day = price_corr(scaled_feats_df, price_2day, interval=2)
small_corr_2day, feat_list_2day = simplified_corr(corr_2day)
small_corr_2day
```

### 3-Day Price Correlations

```python
corr_3day = price_corr(scaled_feats_df, price_3day, interval=3)
small_corr_3day, feat_list_3day = simplified_corr(corr_3day)
small_corr_3day
```

### 7-Day Price Correlations

```python
corr_7day = price_corr(scaled_feats_df, price_7day, interval=7)
small_corr_7day, feat_list_7day = simplified_corr(corr_7day)
small_corr_7day
```

### 14-Day Price Correlations

```python
corr_14day = price_corr(scaled_feats_df, price_14day, interval=14)
small_corr_14day, feat_list_14day = simplified_corr(corr_14day)
small_corr_14day
```

```python
corr_30day = price_corr(scaled_feats_df, price_30day, interval=30)
small_corr_30day, feat_list_30day = simplified_corr(corr_30day)
small_corr_30day
```
